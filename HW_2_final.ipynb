The fields 'reviews_copy1', 'reviews_copy2', 'reviews_copy3', 'reviews_copy4' are indeed exact copies of already existing reviews, therefore these fields are duplicated and can be removed.

**c) Creation of the 'hosts' collection**

```python
# Define a list of fields that are relevant to be included in the "hosts" collection
hosts_fields = [
    'host_about', 'host_has_profile_pic', 'host_id', 'host_identity_verified', 
    'host_is_superhost', 'host_listings_count', 'host_location', 'host_name', 
    'host_neighbourhood', 'host_picture_url', 'host_response_rate', 
    'host_response_time', 'host_thumbnail_url', 'host_total_listings_count', 
    'host_url', 'host_verifications'
]


db.listingsAndReviews_new.aggregate([
    # First stage: Project only the selected fields to minimize data processing
    {"$project": {field: 1 for field in hosts_fields}},
    
    # Second stage: Group documents by 'host_id' to aggregate data for each host
    {"$group": {
        "_id": "$host_id",  
        "host_data": {"$first": "$$ROOT"},  
        "listings": {"$addToSet": "$_id"}  
    }},
    
    # Third stage: Transform the grouped data to match the desired structure of the hosts collection
    {"$addFields": {
        "host_about": {
            "host_name": "$host_data.host_name",
            "description": "$host_data.host_about",
            "host_has_profile_pic": "$host_data.host_has_profile_pic",
            "host_identity_verified": "$host_data.host_identity_verified",
            "host_is_superhost": "$host_data.host_is_superhost",
            "host_listings_count": "$host_data.host_listings_count",
            "host_location": "$host_data.host_location",
            "host_neighbourhood": "$host_data.host_neighbourhood",
            "host_picture_url": "$host_data.host_picture_url",
            "host_response_rate": "$host_data.host_response_rate",
            "host_response_time": "$host_data.host_response_time",
            "host_thumbnail_url": "$host_data.host_thumbnail_url",
            "host_url": "$host_data.host_url",
            "host_verifications": "$host_data.host_verifications"
        }, 
        "REAL_LISTINGS_IDS": "$listings"
    }},
    
    # Fourth stage: Project only the necessary fields for the final output
    {"$project": {
        "_id": 1,
        "host_about": 1, 
        "REAL_LISTINGS_IDS": 1
    }},
    
    # Final stage: Output the results to a new 'hosts' collection
    {"$out": {"db": "sample_airbnb", "coll": "hosts"}}
])
```

**d) Creation of the 'transactions' collection**

```python
# Define the aggregation pipeline to transform data from listingsAndReviews_new collection to a new format in the transactions collection
pipeline = [

    {"$unwind": "$transactions.transactions"},
    # Project specific fields to restructure the document
    {"$project": {
        "_id": 0, 
        "listing_id": "$_id",  
        "date": "$transactions.transactions.date", 
        "price": "$transactions.transactions.price", 
        "bucket_start_date": "$transactions.bucket_start_date",  
        "bucket_end_date": "$transactions.bucket_end_date"  
    }},
    # Output the transformed documents into a new collection named 'transactions'
    {"$out": "transactions"}
]

# This operation will populate the 'transactions' collection with the processed data
db.listingsAndReviews_new.aggregate(pipeline)
```

**e) Show a sample of the latest 6 reviews for every listing so that web users can visualize them**

```python
pipeline = [
    {
        "$lookup": {
            "from": "reviews",  # Specifies the collection in the same database to perform the join
            "localField": "_id",  
            "foreignField": "listing_id",  
            "as": "REVIEWS"  
        }
    },
    {
        "$unwind": {
            "path": "$REVIEWS"  # Deconstructs the 'reviews' array to output a document for each element
        }
    },
    {
        "$sort": {
            "REVIEWS.date": -1  # Sorts the documents by the 'date' field within the embedded 'reviews' array, descending
        }
    },
    {
        "$group": {
            "_id": "$_id",  # Grouping key - the listing ID
            "doc": {"$first": "$$ROOT"},  # Captures the entire document as it appears after the sort (with most recent review first)
            "REVIEWS": {"$push": "$REVIEWS"}  # Collects all reviews into an array
        }
    },
    {
        "$addFields": {
            "doc.REVIEWS": {
                "$slice": ["$REVIEWS", 6]  # Limits the 'reviews' array in the document to the first 6 reviews
            }
        }
    },
    {
        "$replaceRoot": {
            "newRoot": "$doc"  # Replaces the root of the document with the 'doc' field
        }
    },
    {
        "$project": {
            # we will exclude some fields from the output
            "calendar_last_scraped": 0,  
            "last_scraped": 0,
            "first_review": 0,
            "last_review": 0,
            "REVIEWS._id": 0,  
            "REVIEWS.listing_id": 0  
        }
    },
    {
        "$limit": 1  # Limits the output to only one document
    }
]

# Execute the aggregation pipeline
listings_with_reviews = db.listings.aggregate(pipeline)

for listing in listings_with_reviews:
    pprint(listing)
```

### Question 6

We are thinking of trying to assess the security of different areas based on the security deposit required. What is the biggest and smallest difference between the price and deposit for security per number of visitors staying at the property?

```python
# Define an aggregation pipeline to find the biggest and smallest differences between price and security deposit per visitor
pipeline = [
    {
        # Ensure only documents with a non-empty 'security_deposit' are processed
        '$match': {
            'security_deposit': {'$exists': True, '$ne': ''}
        }
    },
    {
        # Add new fields to each document converting price and security deposit to numeric values and ensuring 'accommodates' does not cause division by zero
        '$addFields': {
            'PRICE_NUMERIC': {'$convert': {'input': '$price', 'to': 'double', 'onError': 0}},  # Convert price to double, default to 0 on error
            'DEPOSIT_NUMERIC': {'$convert': {'input': '$security_deposit', 'to': 'double', 'onError': 0}},  # Convert deposit to double, default to 0 on error
            'ACCOMMODATES_NUMERIC': {'$ifNull': ['$accommodates', 1]}  # Use 1 if 'accommodates' is null to avoid division by zero
        }
    },
    {
        # Calculate the absolute difference between price and deposit per visitor
        '$addFields': {
            'DIFFERENCE_PER_VISITOR': {
                '$divide': [
                    {'$abs': {'$subtract': ['$PRICE_NUMERIC', '$DEPOSIT_NUMERIC']}},  # Subtract deposit from price and take absolute value
                    '$ACCOMMODATES_NUMERIC'  # Divide by the number of accommodates to normalize per visitor
                ]
            }
        }
    },
    {
        # Sort documents by the calculated difference in descending order to find the biggest difference first
        '$sort': {'DIFFERENCE_PER_VISITOR': -1}
    },
    {
        # Create two categories in the output: one for the biggest difference and one for the smallest
        '$facet': {
            'BIGGEST_DIFFERENCE': [{'$limit': 1}, {'$project': {'_id': 0, 'name': 1, 'DIFFERENCE_PER_VISITOR': 1}}],  # Limit to 1 document for the biggest difference
            'SMALLEST_DIFFERENCE': [{'$sort': {'DIFFERENCE_PER_VISITOR': 1}}, {'$limit': 1}, {'$project': {'_id': 0, 'name': 1, 'DIFFERENCE_PER_VISITOR': 1}}]  # Sort in ascending and limit to 1 for the smallest
        }
    }
]

# Execute the aggregation pipeline
result = db.listings.aggregate(pipeline)

# Iterate through the results and print the biggest and smallest differences
for r in result:
    print('RESULTS:\n')
    print("Biggest difference per visitor:")
    if r['BIGGEST_DIFFERENCE']:  # Check if there's a result
        print(r['BIGGEST_DIFFERENCE'][0])
    print('\n')
    print("Smallest difference per visitor:")
    if r['SMALLEST_DIFFERENCE']:  # Check if there's a result
        print(r['SMALLEST_DIFFERENCE'][0])
```

What is the average duration of stay (in nights) per type of property per city (you can use the maximum_nights to measure length of stays)? For each property type return the city with the highest and lowest average value.

**Note: Instead of the mean operator it was decided to use the median to calculate the average duration of stay to get results that are robust to outliers. It is noteworthy to mention that it was initially tried with the mean operator, however, unrealistic results were being returned, such as listings presenting about 14 000 years of average maximum nights** 

```python
pipeline = [
    {
        "$match": {
            "maximum_nights": {"$exists": True},
            "address.market": {"$exists": True},  
            "property_type": {"$exists": True}
        }
    },
    {
        "$group": {
            "_id": {"city": "$address.market", "property_type": "$property_type"},
            "maximum_nights": {"$push": {"$toInt": "$maximum_nights"}}
        }
    },
    {
        "$sort": {"_id": 1}
    }
]


# Execute the aggregation pipeline and retrieve results
results = db.listings.aggregate(pipeline)

# Dictionary to store median values by property type for comparison
city_medians = {}

# Process each result from the aggregation
for result in results:
    city = result['_id']['city']
    property_type = result['_id']['property_type']
    nights = result['maximum_nights']
    # Compute the median of maximum nights which is robust against outliers
    median = statistics.median(nights)

    # Initialize the dictionary with property type if not already present
    if property_type not in city_medians:
        city_medians[property_type] = []
    # Append the city and its median maximum nights to the list associated with the property type
    city_medians[property_type].append((city, median))

# Analyze the median data to find the highest and lowest values for each property type
for property_type, data in city_medians.items():
    # Sort the list of tuples containing city and median nights
    data_sorted = sorted(data, key=lambda x: x[1])  
    lowest = data_sorted[0]  
    highest = data_sorted[-1] 
    
    # Print the findings for each property type
    print(f"Property Type: {property_type}")
    print(f"City with the lowest median maximum nights: {lowest[0]} ({lowest[1]} nights)")
    print(f"City with the highest median maximum nights: {highest[0]} ({highest[1]} nights)")
    print("-" * 80)
```

```python
# Create an index for the 'date' field in the transactions collection
# This index, named 'transaction_date_idx', optimizes the performance of queries
# that filter or sort by transaction dates, such as calculating averages over
# specific periods. The '1' specifies an ascending order for the index
db.transactions.create_index([("date", 1)], name="transaction_date_idx")
```

```python
# Prompt the user for input to get the property ID or name and the desired date range
property_identifier = input("Enter the property ID or name: ")
start_year, start_month, start_day = map(int, input("Enter start date (YYYY MM DD): ").split())
end_year, end_month, end_day = map(int, input("Enter end date (YYYY MM DD): ").split())

# Convert the start and end dates to datetime objects for filtering
start_date = datetime(start_year, start_month, start_day)
end_date = datetime(end_year, end_month, end_day)

# Determine if the provided identifier is an ID or a name and find the corresponding property ID if needed
if property_identifier.replace('-', '').isalnum() and not property_identifier.isalpha():
    property_id = property_identifier
else:
    # Look up the property by name to find its ID
    property_doc = db.listings.find_one({"name": property_identifier}, {"_id": 1})
    if property_doc:
        property_id = property_doc['_id']
    else:
        print("Property not found.")
        exit()

# Define an aggregation pipeline to calculate the average value of transactions for the specified property and date range
pipeline = [
    {"$match": {"listing_id": property_id, "date": {"$gte": start_date, "$lte": end_date}}},  # Filter transactions by property ID and date range
    {
        "$group": {
            "_id": None,  # Group by null to aggregate over all matching documents
            "AVERAGE_TRANSACTIONS_VALUE": {"$avg": {"$toDecimal": "$price"}}  # Calculate the average of the transaction prices converted to decimal
        }
    }
]

# Execute the aggregation pipeline
result = list(db.transactions.aggregate(pipeline))

# Output the results
if result and 'AVERAGE_TRANSACTIONS_VALUE' in result[0]:
    avg_value = result[0]['AVERAGE_TRANSACTIONS_VALUE']
    print(f"Property ID: {property_id}, Average Transaction Value: {avg_value}")
else:
    print("No transactions found for the specified period.")
```

We wish to have a summary webpage that displays information about our top destinations. This webpage should display for each of the top 10 cities some basic information about our operations in the area (number of properties by type for example, average price by type) but you can choose the metrics. For each of the top 10 cities it should also provide some basic information about the top 3 properties in each city (price, number of review, whatever you think useful) to show an example of the properties available in the area. We would like to keep this webpage up to date as information changes.

```python
# Define the aggregation pipeline for obtaining the top 10 cities based on total reviews
top_cities_pipeline = [
    {"$group": {
        "_id": "$address.market",  # Group data by city
        "TOTAL_REVIEWS": {"$sum": "$number_of_reviews"}  # Sum up all reviews for each city
    }},
    {"$sort": {"TOTAL_REVIEWS": -1}},  
    {"$limit": 10}  
]

top_cities = list(db.listings.aggregate(top_cities_pipeline))
top_city_names = [city['_id'] for city in top_cities]


for city in top_city_names:
    # Pipeline for city data with average calculations and rounding
    city_data_pipeline = [
        {"$match": {"address.market": city}},
        {"$group": {
            "_id": "$property_type",
            "AVERAGE_PRICE": {"$avg": "$price"},
            "AVERAGE_ACCOMMODATES": {"$avg": "$accommodates"},
            "AVERAGE_RATING": {"$avg": "$METRICS.rating"}
        }},
        {"$project": {
            "AVERAGE_PRICE": {"$round": ["$AVERAGE_PRICE", 3]},
            "AVERAGE_ACCOMMODATES": {"$toInt": {"$round": ["$AVERAGE_ACCOMMODATES"]}},
            "AVERAGE_RATING": {"$round": ["$AVERAGE_RATING", 3]}
        }},
        {"$sort": {"AVERAGE_RATING": -1}},
        {"$limit": 10} 
    ]

    # Pipeline to get top 3 properties in the city based on the number of reviews
    properties_info_pipeline = [
        {"$match": {"address.market": city}},
        {"$sort": {"number_of_reviews": -1}},
        {"$limit": 3},
        {"$project": {
            "name": 1,
            "price": {"$round": ["$price", 3]},  # Round the price to three decimal places
            "number_of_reviews": 1,
            "maximum_nights": 1,
            "minimum_nights": 1,
            "accommodates": {"$toInt": "$accommodates"},  # Convert accommodates directly to integer
            "security_deposit": {"$round": ["$security_deposit", 3]}
        }}
    ]

    # Execute aggregation pipelines for city data and top properties
    city_data = list(db.listings.aggregate(city_data_pipeline))
    top_properties = list(db.listings.aggregate(properties_info_pipeline))

    # Print city data
    print(f"\nCity: {city}")
    for data in city_data:
        print(f"Property Type: {data['_id']}, Avg Price: {data['AVERAGE_PRICE']}, Avg Accommodates: {data['AVERAGE_ACCOMMODATES']}, Avg Rating: {data['AVERAGE_RATING']}")

    # Print top 3 properties
    print("\nTop 3 Properties:")
    for prop in top_properties:
        print(f"Name: {prop['name']}, Price: {prop['price']}, Number of Reviews: {prop['number_of_reviews']}, Max Nights: {prop['maximum_nights']}, Min Nights: {prop['minimum_nights']}, Accommodates: {prop['accommodates']}, Security Deposit: {prop.get('security_deposit', 'N/A')}")
```
